```python
import pandas as pd
import numpy as np

from sklearn import utils
import torch
import torch.nn.functional as F
```

# Preparing Dataset

```python
# laod filtered data
features = np.load('features_f.npy', allow_pickle=True)
labels = np.load('labels_f.npy', allow_pickle=True)
```

```python
# convert features to array
features = features.transpose()
features
```

```python
# skiping first values as in the guide
np_ftrs = features[1:]
np_ftrs
```

```python
len(np_ftrs)
```

```python
# filter values that are not integers
for idx,data in enumerate(np_ftrs[3]):
    if type(data)!= int:
        print(data)
        np_ftrs[3][idx]= 'nan'
np_ftrs
```

```python
# convert strings to numbers
for idx, data in enumerate(np_ftrs):
    np_ftrs[idx]=np.unique(np_ftrs[idx], return_inverse=True)[1]
    print(np_ftrs[idx])
```

```python
# converting nan to numbers
for i in range(len(np_ftrs)):
    try:
        np_ftrs[i] = np_ftrs[i].astype(int)
    except:
        for idx, ele in enumerate(np_ftrs[i]):
            if ele!= 0.0 and ele!= 1.0:
                np_ftrs[i][idx] = 0.0
        np_ftrs[i] = np_ftrs[i].astype(int)
    
```

```python
# swiching to np featues
np_ftrs = np_ftrs[1:]
np_ftrs = np_ftrs.transpose()
np_ftrs
```

```python
df = pd.DataFrame(np_ftrs)
print(df)
# checking what we got  after swiching to np features
```

```python
tr_ftrs = F.normalize(torch.FloatTensor(list(np_ftrs)))
```

```python
tr_ftrs
```

```python
lbls = labels.transpose()
lbls.shape
```

```python
lbls[5]
```

```python
lbl_v = lbls[5]
```

```python
lbl_v
```

```python
lbls[4]
```

```python
lbl_av = lbls[3]
lbl_bv = lbls[4]
```

```python
# setting sucsess and failure values
for idx, data in enumerate(lbl_v):
    if data == '#VALUE!':
        try:
            temp = float(lbl_bv[idx].replace(',',''))/ float(lbl_av[idx].replace(',',''))
            if temp>float(0):
                lbl_v[idx] = float(1)
            else:
                lbl_v[idx]= float(0)
        except:
            lbl_v[idx] = float(0)
    elif data == '#DIV/0!':
        try:
            temp = float(lbl_bv[idx].replace(',',''))/ float(lbl_av[idx].replace(',',''))
            if temp>float(0):
                lbl_v[idx] = float(1)
            else:
                lbl_v[idx] = float(0)
        except:
            lbl_v[idx] = float(0)
    elif float(data.replace(',','')) > float(0):
        lbl_v[idx] = float(1)
    else:
        lbl_v[idx] = float(0)
        

lbl_v = lbl_v.astype(float)
```

```python
tr_lbl = torch.LongTensor(lbl_v)
# changing the values to long tensor to work with the model
```

```python
tr_lbl
```

```python
#shuffle
tr_ftrs, tr_lbl= utils.shuffle(tr_ftrs, tr_lbl)
```

```python
print(len(tr_ftrs), len(tr_lbl))
```

```python
# split data to train test
train_ftrs = tr_ftrs[0:646]
train_lbl = tr_lbl[0:646]

test_ftrs = tr_ftrs[646:808]
test_lbl = tr_lbl[646:808]

print(len(train_ftrs), len(test_ftrs))
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression


# Create and fit the linear regression model
model = LinearRegression()
model.fit(train_ftrs, train_lbl)

# Print the coefficients and intercept
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

# Predict on new data
predictions = model.predict(test_ftrs)
print("Predictions:", predictions)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Create and fit the linear regression model
model = LinearRegression()
model.fit(train_ftrs, train_lbl)

# Predict on the test set
predictions = model.predict(test_ftrs)

# Calculate evaluation metrics
mae = mean_absolute_error(test_lbl, predictions)
mse = mean_squared_error(test_lbl, predictions)
rmse = np.sqrt(mse)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
from sklearn.metrics import accuracy_score

# Assuming you have already trained the model and obtained predictions
predicted_labels = model.predict(test_ftrs)  # Predicted labels for the test set

# Convert predicted probabilities to binary labels (if needed)
predicted_labels = (predicted_labels > 0.5).astype(int)

# Calculate accuracy
accuracy = accuracy_score(test_lbl, predicted_labels)

print("Accuracy:", accuracy)
```

```python
from sklearn.linear_model import Ridge, Lasso

# Create and fit the Ridge regression model
ridge_model = Ridge(alpha=0.5)  # alpha is the regularization strength
ridge_model.fit(train_ftrs, train_lbl)

# Create and fit the Lasso regression model
lasso_model = Lasso(alpha=0.5)  # alpha is the regularization strength
lasso_model.fit(train_ftrs, train_lbl)

# Predict on the test set using Ridge and Lasso models
ridge_predictions = ridge_model.predict(test_ftrs)
lasso_predictions = lasso_model.predict(test_ftrs)

# Calculate accuracy or other evaluation metrics for Ridge and Lasso predictions
ridge_accuracy = accuracy_score(test_lbl, (ridge_predictions > 0.5).astype(int))
lasso_accuracy = accuracy_score(test_lbl, (lasso_predictions > 0.5).astype(int))

print("Ridge Accuracy:", ridge_accuracy)
print("Lasso Accuracy:", lasso_accuracy)
```

```python
from sklearn.linear_model import ElasticNet

# Create and fit the Elastic Net model
elastic_net_model = ElasticNet(alpha=0.5, l1_ratio=0.5)  # alpha and l1_ratio are regularization parameters
elastic_net_model.fit(train_ftrs, train_lbl)

# Predict on the test set using the Elastic Net model
elastic_net_predictions = elastic_net_model.predict(test_ftrs)

# Calculate accuracy or other evaluation metrics for Elastic Net predictions
elastic_net_accuracy = accuracy_score(test_lbl, (elastic_net_predictions > 0.5).astype(int))

print("Elastic Net Accuracy:", elastic_net_accuracy)
```
