```python
import pandas as pd
```

```python
#loading data
features = pd.read_csv('startup_founder_chars.csv',encoding='latin-1')
lbls = pd.read_csv('startup_valuations.csv',encoding='latin-1')
```

```python
#checking the columns
features.columns
```

```python
lbls.columns
```

```python
#selecting features
#droping featuers with low correlation with labels or bad features proved to be unstable with other features via experimentation
features = features.drop(['Standardized University'], axis=1)
features = features.drop(['Standardized Major'], axis=1)
features = features.drop(['Degree Type'], axis=1)
features = features.drop(['Standardized Graduate Studies'], axis=1)
features = features.drop(['Graduate Diploma'], axis=1)
features = features.drop(['Years of Employment'], axis=1)
features = features.drop(['Worked as product manager/director/head/VP?'], axis=1)
features = features.drop(['Worked at Google?'], axis=1)
features = features.drop(['Worked at Microsoft?'], axis=1)
features = features.drop(['Crunchbase'], axis=1)
features = features.drop(['LinkedIn'], axis=1)
```

```python
features
```

```python
lbls
```

```python
#convert features into numpy array from pandas dataframe
import numpy
np_f = numpy.array(features)
np_l = numpy.array(lbls)
```

```python
#transpose so that instead of every row as an array, every column becomes an array. so that we can loop through all the names to find the common names in the label file.
np_f = np_f.transpose()
np_l = np_l.transpose()
```

```python
#arrays to save the final features and labels in
ftr = []
lbl = []
```

```python
#the first column of np_f and np_l is the 'Full name' column
for ser, name in enumerate(np_l[0]): #loop through every name in label_file
    for idx, data in enumerate(np_f[0]): #for every name in label file, loop through every name in feature file
        if name==data: #if the name in label file is equal to the name in feature file
            ftr.append(np_f.transpose()[idx]) #save that feature file's row in which that name is
            lbl.append(np_l.transpose()[ser]) #also save the label row of that name
            break #no need to go through more names, we have found our common row
    
```

```python
numpy.save('features_f',ftr, allow_pickle=True) #save the extracted
numpy.save('labels_f',lbl, allow_pickle=True) 
```
